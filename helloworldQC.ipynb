```python
from qiskit import *
#Imports qiskit 
from qiskit.visualization import circuit_drawer
#Imports qiskit visualisation tools
import matplotlib.pyplot as plt
#Imports mpl for graphically displaying the circuit. This is one of multiple ways to do so, including the use of statevectors and bloch spheres
from qiskit_aer.aerprovider import AerSimulator 
#Imports Aer simulator to get simulated results for the circuit
from qiskit.visualization import plot_histogram
#Will allow us to use a histogram to display our results later 
from qiskit.visualization import *
from qiskit import result
```

```python
qr = QuantumRegister (2)
cr = ClassicalRegister (2)
#Creates a quantum and classical register, with 2 qubits and 2 bits respectively
#Note that the classical register is needed to store the measurements of the quantum register
circuit = QuantumCircuit (qr, cr)
#Creates the quantum circuit with the two registers
circuit.h(0)
#Applies the Hadamard gate onto the first qubit to bring it into a state of superposition
circuit.cx(0, 1)
#Applies the Control X (also known as the Control NOT gate) to create entanglement between the two qubits; the q_0 is the control qubit and q_1 is the target qubit
%matplotlib inline
circuit.draw(output = "mpl")
#Uses MPL to display the quantum circuit
```

```python
circuit.measure(qr, cr)
#Measures the values of the quantum register and stores them in the classical register
```

```python
simulator = AerSimulator()
#Allows us to use a simulator to test and see our measurements
```

```python
job = simulator.run(circuit) 
#Executes the experiment
```

```python
result = job.result()
#Gets the results from the experiment
```

```python
result.get_counts() 
#Interprets the results as a count dictionary 
```

```python
plot_histogram(result.get_counts())
#A visual respresentation of the above results. Note that due to quantum indeterminacy, the counts are not exactly equal
```
